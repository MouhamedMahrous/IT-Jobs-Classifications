```python
DATA_PATH = "../dataset/Cleand_Data.pkl"
MODEL_PATH = "../models/RandomForest.h5"
SCALER_PATH = "../models/Scaler.h5"
```

```python
import pandas as pd
import joblib
```

```python
df = pd.read_pickle(DATA_PATH)
df.head()
```

```python
x = df.drop("DevType",axis=1)
skills = x.columns.droplevel(0).to_list()
skills
```

```python
dictionary_record = {}
for skill in skills:
    dictionary_record[skill] = 0
dictionary_record    
```

# Now, testing new record by passing it to our trained model

```python
new_record = ['Pandas', 'TensorFlow', 'Torch/PyTorch', 'Python', 'Keras']
for i in new_record:
    dictionary_record[i] = 1

dictionary_record
```

```python
new_record = list(dictionary_record.values())
new_record
```

```python
model = joblib.load(MODEL_PATH)
scaler = joblib.load(SCALER_PATH)
```

```python
model.predict(scaler.transform([new_record]))
```

```python
predictions = model.predict_proba(scaler.transform([new_record]))
posative_probs = [prob[0][1] for prob in predictions]
pd.Series(posative_probs, index=df["DevType"].columns).sort_values(ascending=False)
```
