```python
DATA_PATH = "../dataset/Cleand_Data.pkl"

MULTI_TARGET_MODELS = {
    "KNN" : KNeighborsClassifier(),
    "DT" : DecisionTreeClassifier(),
    "RF" : RandomForestClassifier()
}
NON_MULTI_TARGET_MODELS ={
    "LR" : LogisticRegression(),
    "SVM" : SVC(),
    "NB" : GaussianNB()
}
```

```python
import pandas as pd

import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

# Functions

```python
def calculate_quality(truth, prediction):
    prediction = pd.DataFrame(data=prediction, columns=truth.columns)
    quality_scors = {}
    for col in prediction.columns:
        quality_scors[col] = [accuracy_score(truth[col],prediction[col])*100,
                              precision_score(truth[col],prediction[col])*100,
                              recall_score(truth[col],prediction[col])*100,
                              f1_score(truth[col],prediction[col])*100]
    quality_scors = pd.DataFrame(data = quality_scors, index=["accuracy_score","precision_score","recall_score","f1_score"]).transpose()
    return quality_scors
```

```python
df = pd.read_pickle(DATA_PATH)
df
```

```python
x = df.drop("DevType", axis=1)
y = df["DevType"]
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
x_train.shape
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

```python
for name,model in MULTI_TARGET_MODELS.items():
    print(f"Using: {name} Algoritm")
    print("-"*30)
    print("-"*30)
    model.fit(x_train, y_train)
    y_test_pred = model.predict(x_test)
    y_train_pred = model.predict(x_train)
    print(f"\nEvaluate on training set:")
    print("-"*30)
    scores = calculate_quality(y_train, y_train_pred)
    print(scores)
    print("\n")
    print(scores.mean())
    print(f"\nEvaluate on testing set:")
    print("-"*30)
    scores = calculate_quality(y_test, y_test_pred)
    print(scores)
    print(scores.mean())
    print("-"*100)
```

```python
for name,cla in NON_MULTI_TARGET_MODELS.items():
    print(f"Using: {name} Algoritm")
    print("-"*30)
    print("-"*30)
    model = MultiOutputClassifier(cla)
    model.fit(x_train, y_train)
    y_test_pred = model.predict(x_test)
    y_train_pred = model.predict(x_train)
    print(f"\nEvaluate on training set:")
    print("-"*30)
    scores = calculate_quality(y_train, y_train_pred)
    print(scores)
    print("\n")
    print(scores.mean())
    print(f"\nEvaluate on testing set:")
    print("-"*30)
    scores = calculate_quality(y_test, y_test_pred)
    print(scores)
    print(scores.mean())
    print("-"*100)
```

## After comparing the quality scores of the algoritms:
## Observation :----> RF algorithm will be selected (heighest accuracy & heighest f1_score)

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```
